```python
#Making list of values from label column and checking to see if number of items is correct
list_label = df0['label'].values.tolist()
print(len(list_label))

#Making individual lists based on each rater's labels (which come in a sequence of 3 steps) and checking to see if number of items is correct
rater14 = list_label[0:1500:3]
rater13 = list_label[1:1500:3]
rater11 = list_label[2:1500:3]

print(len(rater14))
print(len(rater13))
print(len(rater11))
```

```python
#Calculating agreement

from sklearn.metrics import cohen_kappa_score
from nltk import agreement

print("Cohen's Kappa between rater14 and rater13:", cohen_kappa_score(rater14, rater13))
print("Cohen's Kappa between rater14 and rater11:", cohen_kappa_score(rater14, rater11))
print("Cohen's Kappa between rater13 and rater11:", cohen_kappa_score(rater13, rater11))

formatted_codes = [[1,i,rater14[i]] for i in range(len(rater14))] + [[2,i,rater13[i]] for i in range(len(rater13))]  + [[3,i,rater11[i]] for i in range(len(rater11))]

ratingtask = agreement.AnnotationTask(data=formatted_codes)

print("Fleiss's Kappa:",ratingtask.multi_kappa())
print("Krippendorff's alpha:",ratingtask.alpha())
print("Scott\'s pi:",ratingtask.pi())
```

```python
data1 = pd.read_csv ("validation_4chan_310521.csv") #this one works
df1 = pd.DataFrame(data1, columns = ["label","user"])
```

```python
idx_15 = data1["user"] == 15
idx_14 = data1["user"] == 14
idx_13 = data1["user"] == 13
idx_11 = data1["user"] == 11
```

```python
user15 = data1.loc[idx_15,:]
user14 = data1.loc[idx_14,:]
user13 = data1.loc[idx_13,:]
user11 = data1.loc[idx_11,:]

user15_list = user15["label"].values.tolist()
user14_list = user14["label"].values.tolist()
user13_list = user13["label"].values.tolist()
user11_list = user11["label"].values.tolist()
```

```python
from sklearn.metrics import cohen_kappa_score
from nltk import agreement

print("Cohen's Kappa between me and Bjørn:", cohen_kappa_score(user11_list, user13_list))
print("Cohen's Kappa between me and Simon:", cohen_kappa_score(user11_list, user14_list))
print("Cohen's Kappa between me and Emily:", cohen_kappa_score(user11_list, user15_list))


formatted_codes = [[1,i,user11_list[i]] for i in range(len(user11_list))] + [[2,i,user13_list[i]] for i in range(len(user13_list))]  + [[3,i,user14_list[i]] for i in range(len(user14_list))] + [[4,i,user15_list[i]] for i in range(len(user15_list))]

ratingtask = agreement.AnnotationTask(data=formatted_codes)

print("Fleiss's Kappa:",ratingtask.multi_kappa())
print("Krippendorff's alpha:",ratingtask.alpha())
print("Scott's pi:",ratingtask.pi())
```

```python
df2 = pd.DataFrame(data1, columns = ["label","user", "text"])
pd.set_option('display.max_colwidth', None)
```

```python
user11_label20 = df2.loc[(df2['user'] == 11) & (df2['label'] == 20)]
user11_label21 = df2.loc[(df2['user'] == 11) & (df2['label'] == 21)]

user13_label20 = df2.loc[(df2['user'] == 13) & (df2['label'] == 20)]
user13_label21 = df2.loc[(df2['user'] == 13) & (df2['label'] == 21)]

user14_label20 = df2.loc[(df2['user'] == 14) & (df2['label'] == 20)]
user14_label21 = df2.loc[(df2['user'] == 14) & (df2['label'] == 21)]

user15_label20 = df2.loc[(df2['user'] == 15) & (df2['label'] == 20)]
user15_label21 = df2.loc[(df2['user'] == 15) & (df2['label'] == 21)]
```

```python
user11_13_label20 = pd.concat([user11_label20, user13_label20])
user11_13_label21 = pd.concat([user11_label21, user13_label21])

user11_14_label20 = pd.concat([user11_label20, user14_label20])
user11_14_label21 = pd.concat([user11_label21, user14_label21])

user11_15_label20 = pd.concat([user11_label20, user15_label20])
user11_15_label21 = pd.concat([user11_label21, user15_label21])
```

```python
#history-comment disc by Bjørn
label20_user11_13_disc = user11_13_label20.drop_duplicates(subset=['text'], keep = False)
idx_label_20_user_13_disc = label20_user11_13_disc['user'] == 13
label20_user_13_disc = label20_user11_13_disc[idx_label_20_user_13_disc]

#non-history-comment disc by Bjørn
label21_user11_13_disc = user11_13_label21.drop_duplicates(subset=['text'], keep = False)
idx_label21_user_13_disc = label21_user11_13_disc['user'] == 13
label21_user_13_disc = label21_user11_13_disc[idx_label21_user_13_disc]

#history-comment disc by Simon
label20_user11_14_disc = user11_14_label20.drop_duplicates(subset=['text'], keep = False)
idx_label20_user_14_disc = label20_user11_14_disc['user'] == 14
label20_user_14_disc = label20_user11_14_disc[idx_label20_user_14_disc]

#non-history-comment disc by Simon
label21_user11_14_disc = user11_14_label21.drop_duplicates(subset=['text'], keep = False)
idx_label21_user_14_disc = label21_user11_14_disc['user'] == 14
label21_user_14_disc = label21_user11_14_disc[idx_label21_user_14_disc]

#history-comment disc by Emily
label20_user11_15_disc = user11_15_label20.drop_duplicates(subset=['text'], keep = False)
idx_label20_user_15_disc = label20_user11_15_disc['user'] == 15
label20_user_15_disc = label20_user11_15_disc[idx_label20_user_15_disc]

#history-comment disc by Emily
label21_user11_15_disc = user11_15_label21.drop_duplicates(subset=['text'], keep = False)
idx_label21_user_15_disc = label21_user11_15_disc['user'] == 15
label21_user_15_disc = label21_user11_15_disc[idx_label21_user_15_disc]
```

```python
#disc by Bjørn
user_13_disc = pd.concat([label20_user_13_disc, label21_user_13_disc])

#disc by Simon
user_14_disc = pd.concat([label20_user_14_disc, label21_user_14_disc])

#disc by Emily
user_15_disc = pd.concat([label20_user_15_disc, label21_user_15_disc])
```
